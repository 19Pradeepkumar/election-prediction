```python
import pandas as pd
import re
import demoji

def contains_emoji(text):
    # Checking if the text contains any emojis
    return bool(demoji.findall(text))

def clean_text(text):
    # Print the first five lines of the CSV file before normalization
    if clean_text.counter < 5:
        print("Before normalization:")
        print(text)
        clean_text.counter += 1

    # Removing emojis
    text = demoji.replace(text, '')

    # Removing white spaces
    text = ' '.join(text.split())

    # Remove special characters excluding Telugu characters
    text = re.sub(r'[^a-zA-Z0-9\s\u0C00-\u0C7F]', '', text)

    # Print the first five lines of the CSV file after normalization
    if clean_text.after_counter < 5:
        print("After normalization:")
        print(text)
        clean_text.after_counter += 1

    return text

def clean_csv(input_file, output_file):
    clean_text.counter = 0
    clean_text.after_counter = 0

    df = pd.read_csv(input_file, encoding='utf-8')  # Make sure to specify the correct encoding if needed

    df['has_emoji'] = df['english_translation'].apply(contains_emoji)

    # Filter out rows with emojis
    df_cleaned = df[df['has_emoji'] == True].copy()

    # Applying text cleaning
    df_cleaned['cleaned_text'] = df_cleaned['english_translation'].apply(clean_text)

    # Removing rows where 'Comment' column is empty after cleaning
    df_cleaned = df_cleaned[df_cleaned['cleaned_text'].str.strip() != '']

    # Saving only the cleaned sentences to the output file
    df_cleaned[['cleaned_text']].to_csv(output_file, index=False, encoding='utf-8')  # Make sure to specify the correct encoding if needed

# Example usage:
input_csv_file = 'bjptel.csv'
output_csv_file = 'bjpe.csv'
clean_csv(input_csv_file, output_csv_file)
```

```python
import pandas as pd

def check_missing_values(file_path):
    df = pd.read_csv(file_path)
    missing_values = df.isnull().sum()
    missing_columns = [column for column in missing_values.index if missing_values[column] > 0]
    return missing_columns

missing_columns_bjp = check_missing_values('bjp.csv')
missing_columns_cong = check_missing_values('cong.csv')

if 'comment' in missing_columns_bjp or 'lemmatized' in missing_columns_bjp:
    print("Missing values found in bjp.csv")
else:
    print("No missing values found in bjp.csv")

if 'comment' in missing_columns_cong or 'lemmatized' in missing_columns_cong:
    print("Missing values found in cong.csv")
else:
    print("No missing values found in cong.csv")
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

# List of negation words
negation_words = ['not', 'no', 'never', 'none', 'neither', 'nor', 'nowhere', 'hardly', 'scarcely', 'barely']

# Sample custom lexicon (you can replace this with your own custom lexicon)
custom_lexicon = {'good': 0.5, 'bad': -0.5, 'great': 0.8, 'terrible': -0.8}

def adjust_sentiment_score(score, negated):
    if negated:
        # Double the positive polarity for words following negation
        return 2 * score if score > 0 else score
    else:
        return score

def analyze_sentiment(comment):
    sia = SentimentIntensityAnalyzer()

    # Initialize sentiment scores
    custom_score = 0
    vader_score = sia.polarity_scores(comment)['compound']

    # Handling negation
    negated = False
    for word in comment.split():
        if word.lower() in negation_words:
            negated = not negated  # Toggle the negation flag
        else:
            # Adjust sentiment scores based on negation and invert polarity if negation is present
            custom_score += adjust_sentiment_score(custom_lexicon.get(word, 0), negated)

    # Combine scores or give priority to the custom lexicon
    final_score = 0.7 * vader_score + 0.3 * custom_score

    if final_score >= 0.05:
        return 'positive'
    elif final_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

def predict_election_result(comments):
    positive_count = 0
    negative_count = 0

    for _, comment in comments.iterrows():
        final_sentiment = analyze_sentiment(comment['lemmatized'])

        if final_sentiment == 'positive':
            positive_count += 1
        elif final_sentiment == 'negative':
            negative_count += 1
    return positive_count+negative_count

# Read comments from CSV file with headers
def read_comments_from_csv(file_path):
    df = pd.read_csv(file_path, encoding='utf-8')
    # Fill missing values in 'lemmatized' column with empty string
    df['lemmatized'] = df['lemmatized'].fillna('')
    # Convert 'lemmatized' column to string type
    df['lemmatized'] = df['lemmatized'].astype(str)
    return df[['comment', 'lemmatized']]


csv_file_path1 = 'bjp.csv'
csv_file_path2 = 'cong.csv'
social_media_comments1 = read_comments_from_csv(csv_file_path1)
social_media_comments2 = read_comments_from_csv(csv_file_path2)

bjp_prediction = predict_election_result(social_media_comments1)
congress_prediction = predict_election_result(social_media_comments2)
print("BJP Score: ",bjp_prediction,"\nCongress Score: ",congress_prediction)
if bjp_prediction > congress_prediction:
    print( "BJP party is predicted to win")
elif congress_prediction>bjp_prediction:
    print("Congress party is predicted to win")
else:
    print("It'sa tie or inconclusive")
```

```python
import pandas as pd
from googletrans import Translator
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk
from nltk.corpus import wordnet

# Downloading NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

def translate_telugu_to_english(text):
    translator = Translator()
    translation = translator.translate(text, src='te', dest='en')
    return translation.text.lower()

def perform_stemming_and_lemmatization(text):
    # Tokenize the sentence
    tokens = nltk.word_tokenize(text)

    # Performing stemming
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]

    # Performing lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(tag)) for token, tag in nltk.pos_tag(tokens) if get_wordnet_pos(tag) is not None]

    # Combining the stemmed and lemmatized tokens into a string
    stemmed_text = ' '.join(stemmed_tokens)
    lemmatized_text = ' '.join(lemmatized_tokens)

    return stemmed_text, lemmatized_text

def get_wordnet_pos(tag):
    if tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    elif tag.startswith('J'):
        return wordnet.ADJ
    # Remove the default to noun for unidentified words
    # else:
    #     return wordnet.NOUN

def translate_stem_and_lemmatize_csv(input_file, output_file):
    # Read CSV without header
    df = pd.read_csv(input_file, header=None, names=['comment'], encoding='utf-8')

    df['english_translation'] = df['comment'].apply(translate_telugu_to_english)

    # Perform stemming and lemmatization
    df['stemmed'], df['lemmatized'] = zip(*df['english_translation'].apply(perform_stemming_and_lemmatization))

    # Filter out rows with empty translations
    df_cleaned = df[df['english_translation'].str.strip() != ''].copy()

    # Save the results to the output file
    df_cleaned[['comment', 'english_translation', 'stemmed', 'lemmatized']].to_csv(output_file, index=False, encoding='utf-8')

input_csv_file = 'bjptel.csv'
output_csv_file = 'bjpe.csv'
translate_stem_and_lemmatize_csv(input_csv_file, output_csv_file)
```
